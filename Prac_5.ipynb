```python
import math
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from pandas import DataFrame, Series
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics


class Trigonometric:
    @staticmethod
    def cos(x):
        return math.cos(x)

    @staticmethod
    def sin(x):
        return math.sin(x)

    @staticmethod
    def tan(x):
        return Trigonometric.sin(x) / Trigonometric.cos(x)

    @staticmethod
    def asin(x):
        return math.asin(x)

    @staticmethod
    def acos(x):
        return math.acos(x)

    @staticmethod
    def atan(x):
        return Trigonometric.asin(x) / Trigonometric.acos(x)

    @staticmethod
    def degrees_to_radian(x):
        return x * math.pi / 180


class Tree:
    end = False

    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Tree(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Tree(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def PrintRight(self):
        print(self.data)
        if self.right is not None:
            self.right.PrintRight()

    def PrintLeft(self):
        if self.left is not None and self.end == False:
            self.left.PrintLeft()
        else:
            self.end = True
            print(self.data)
            if self.right is not None:
                self.right.PrintLeft()

    def PrintTree(self):
        self.PrintLeft()
        self.PrintRight()


def task_1():
    print(Trigonometric.cos(1),
    Trigonometric.sin(3.14),
    Trigonometric.degrees_to_radian(180))


def task_2():
    T = ["a", ["b", ["d", "e"]], ["c", ["f"]]]
    print(T[0])
    print(T[1])
    print(T[2])


def task_3():
    root = Tree(6)
    for i in range(0, 10):
        root.insert(i)
    root.PrintTree()


def task_4():
    x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
    target = [0, 0, 0, 1, 1, 1]
    dsx = pd.DataFrame(x)
    dsy = pd.DataFrame(target)
    print(dsx)
    print(dsx.shape)
    print(dsx.head())
    x_train, x_test, y_train, y_test = train_test_split(
        dsx,
        dsy,
        test_size=0.15
    )

    x_train.head()
    y_train.head()

    classifier = DecisionTreeClassifier()
    classifier.fit(x_train, y_train)
    tree.plot_tree(classifier)
    y_predict = classifier.predict(x_test)
    print(y_predict)
    #print(confusion_matrix(y_test, y_predict))
    print(classification_report(y_test, y_predict))


def task_5():
    url = r'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
    df = pd.read_csv(url)
    df.head()
    print(df.shape)
    df.describe()
    print(df)
    plt.scatter(df['Petrol_tax'], df['Petrol_Consumption'], color='b', label="exp")
    plt.xlabel("Налог на нефть")
    plt.ylabel("Потребление нефти")
    plt.show()
    plt.scatter(df['Average_income'], df['Petrol_Consumption'], color='b', label="exp")
    plt.xlabel("Средняя зарплата")
    plt.ylabel("Потребление нефти")
    plt.show()
    plt.scatter(df['Paved_Highways'], df['Petrol_Consumption'], color='b', label="exp")
    plt.xlabel("Количество сделанных дорог")
    plt.ylabel("Потребление нефти")
    plt.show()
    plt.scatter(df['Population_Driver_licence(%)'], df['Petrol_Consumption'], color='b', label="exp")
    plt.xlabel("Количество людей с водительскими правами")
    plt.ylabel("Потребление нефти")
    plt.show()
    x = df[['Average_income', 'Population_Driver_licence(%)', 'Petrol_Consumption', 'Petrol_tax']]
    y = df['Petrol_Consumption']
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
    reg = DecisionTreeRegressor()
    reg.fit(x_train, y_train)
    y_predict = reg.predict(x_test)
    print(y_predict)
    ds = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict})
    print(ds)
    print('MSError:', metrics.mean_squared_error(y_test, y_predict))
    print('MAError:', metrics.mean_absolute_error(y_test, y_predict))
    print('DecisionTreeRegressor')
    tree.plot_tree(reg)


tasks = [task_1, task_2, task_3, task_4, task_5]
tasks[int(input("Input task number: ")) - 1]()






```
